```python
import tkinter as Tk
from tkinter import *
from tkinter import ttk
import mysql.connector
from tkinter import messagebox
```

```python
class HotelManagementSystem:
    def __init__(self,root):
        self.root=root
        self.root.title("Hotel Management System")
        self.root.geometry("1550x800+0+0")



        self.USER_var=StringVar()
        self.CUSTOMERID_var=StringVar()
        self.FIRST_NAME_var=StringVar()
        self.LAST_NAME_var=StringVar()
        self.MOB_var=StringVar()
        self.BOOKING_TYPE_var=StringVar()
        self.CHECK_IN_var=StringVar()
        self.CHECK_OUT_var=StringVar()
        self.CHARGE_var=StringVar()
        self.QUANTITY_var=StringVar()
        self.TOTAL_AMOUNT_var=StringVar()
        self.VERIFICATIONID_var=StringVar()

        lbltitle=Label(self.root,text="Hotel Management System",bg="powder blue",fg="green",bd=20,relief=RIDGE,font=("times new roman",50,"bold"),padx=2,pady=6)
        lbltitle.pack(side=TOP,fill=X)

        frame=Frame(self.root,bd=12,relief=RIDGE,padx=20,bg="powder blue")
        frame.place(x=0,y=130,width=1530,height=400)
        
        DataFrame=LabelFrame(frame,text="Hotel Management information",bg="powder blue",fg="green",bd=12,relief=RIDGE,font=("times new roman",12,"bold"))
        DataFrame.place(x=0,y=5,width=900,height=350)
        
        lblUSER=Label(DataFrame,bg="powder blue",text="USER Type",font=("arial",12,"bold"),padx=2,pady=6)
        lblUSER.grid(row=0,column=0,sticky=W)
        

        comUSER=ttk.Combobox(DataFrame,font=("arial",12,"bold"),textvariable=self.USER_var,width=27,state="readonly")
        comUSER["value"]=("Owner","Staff","Manager")
        comUSER.grid(row=0,column=1)

        lblCUSTOMERID=Label(DataFrame,bg="powder blue",text="CUSTOMERID:",font=("arial",12,"bold"),padx=2,pady=6)
        lblCUSTOMERID.grid(row=1,column=0,sticky=W)
        txtCUSTOMERID=Entry(DataFrame,font=("arial",13,"bold"),textvariable=self.CUSTOMERID_var,width=29)
        txtCUSTOMERID.grid(row=1,column=1)

    
        lblFIRST_NAME=Label(DataFrame,bg="powder blue",text="FIRST_NAME:",font=("arial",12,"bold"),padx=2,pady=6)
        lblFIRST_NAME.grid(row=2,column=0,sticky=W)
        txtFIRST_NAME=Entry(DataFrame,font=("arial",13,"bold"),textvariable=self.FIRST_NAME_var,width=29)
        txtFIRST_NAME.grid(row=2,column=1)

        lblLAST_Name=Label(DataFrame,bg="powder blue",text="LAST_Name:",font=("arial",12,"bold"),padx=2,pady=6)
        lblLAST_Name.grid(row=3,column=0,sticky=W)
        txtLAST_Name=Entry(DataFrame,font=("arial",13,"bold"),textvariable=self.LAST_NAME_var,width=29)
        txtLAST_Name.grid(row=3,column=1)
        
        lblMOB=Label(DataFrame,bg="powder blue",text="MOB:",font=("arial",12,"bold"),padx=2,pady=6)
        lblMOB.grid(row=4,column=0,sticky=W)
        txtMOB=Entry(DataFrame,font=("arial",13,"bold"),textvariable=self.MOB_var,width=29)
        txtMOB.grid(row=4,column=1)
        
        lblBOOKING_TYPE=Label(DataFrame,bg="powder blue",text="BOOKING_TYPE:",font=("arial",12,"bold"),padx=2,pady=6)
        lblBOOKING_TYPE.grid(row=5,column=0,sticky=W)

        comBOOKING_TYPE=ttk.Combobox(DataFrame,font=("arial",12,"bold"),textvariable=self.BOOKING_TYPE_var,width=27,state="readonly")
        comBOOKING_TYPE["value"]=("AC","NON-AC","DELUX")
        comBOOKING_TYPE.grid(row=5,column=1)
        
        lblCHECK_IN=Label(DataFrame,bg="powder blue",text="CHECK_IN:",font=("arial",12,"bold"),padx=2,pady=6)
        lblCHECK_IN.grid(row=6,column=0,sticky=W)
        txtCHECK_IN=Entry(DataFrame,font=("arial",13,"bold"),textvariable=self.CHECK_IN_var,width=29)
        txtCHECK_IN.grid(row=6,column=1)
        
        lblCHECK_OUT=Label(DataFrame,bg="powder blue",text="CHECK_OUT:",font=("arial",12,"bold"),padx=2,pady=6)
        lblCHECK_OUT.grid(row=7,column=0,sticky=W)
        txtCHECK_OUT=Entry(DataFrame,font=("arial",13,"bold"),textvariable=self.CHECK_OUT_var,width=29)
        txtCHECK_OUT.grid(row=7,column=1)
    
        lblCHARGE=Label(DataFrame,bg="powder blue",text="CHARGE:",font=("arial",12,"bold"),padx=2,pady=6)
        lblCHARGE.grid(row=8,column=0,sticky=W)
        txtCHARGE=Entry(DataFrame,font=("arial",13,"bold"),textvariable=self.CHARGE_var,width=29)
        txtCHARGE.grid(row=8,column=1)
        
        lblQUANTITY=Label(DataFrame,bg="powder blue",text="QUANTITY:",font=("arial",12,"bold"),padx=2,pady=6)
        lblQUANTITY.grid(row=0,column=2,sticky=W)
        txtQUANTITY=Entry(DataFrame,font=("arial",13,"bold"),textvariable=self.QUANTITY_var,width=29)
        txtQUANTITY.grid(row=0,column=3)
        
        lblTOTAL_AMOUNT=Label(DataFrame,bg="powder blue",text="TOTAL_AMOUNT:",font=("arial",12,"bold"),padx=2,pady=6)
        lblTOTAL_AMOUNT.grid(row=1,column=2,sticky=W)
        txtTOTAL_AMOUNT=Entry(DataFrame,font=("arial",13,"bold"),textvariable=self.TOTAL_AMOUNT_var,width=29)
        txtTOTAL_AMOUNT.grid(row=1,column=3)
        
        lblVERIFICATIONID=Label(DataFrame,bg="powder blue",text="VERIFICATIONID:",font=("arial",12,"bold"),padx=2,pady=6)
        lblVERIFICATIONID.grid(row=2,column=2,sticky=W)

        comVERIFICATIONID=ttk.Combobox(DataFrame,font=("arial",12,"bold"),textvariable=self.VERIFICATIONID_var,width=27,state="readonly")
        comVERIFICATIONID["value"]=("AADHAR","PAN","OTHER")
        comVERIFICATIONID.grid(row=2,column=3)
        


        framebutton=Frame(self.root,bd=12,relief=RIDGE,padx=20,bg="powder blue")
        framebutton.place(x=0,y=530,width=1530,height=70)
        
        
        btnAddData=Button(framebutton,command=self.add_data,text="Add Data",font=("arial",12,"bold"),width=23,bg="blue",fg="white")
        btnAddData.grid(row=0,column=0)
        
        btnAddData=Button(framebutton,command=self.showData,text="Show Data",font=("arial",12,"bold"),width=23,bg="blue",fg="white")
        btnAddData.grid(row=0,column=1)
        
        btnAddData=Button(framebutton,command=self.update,text="Update",font=("arial",12,"bold"),width=23,bg="blue",fg="white")
        btnAddData.grid(row=0,column=2)
        
        btnAddData=Button(framebutton,command=self.delete,text="Delete",font=("arial",12,"bold"),width=23,bg="blue",fg="white")
        btnAddData.grid(row=0,column=3)
        
        btnAddData=Button(framebutton,command=self.reset,text="Reset",font=("arial",12,"bold"),width=23,bg="blue",fg="white")
        btnAddData.grid(row=0,column=4)
        
        btnAddData=Button(framebutton,command=self.iExit,text="Exit",font=("arial",12,"bold"),width=23,bg="blue",fg="white")
        btnAddData.grid(row=0,column=5)
        
        FrameDetails=Frame(self.root,bd=12,relief=RIDGE,padx=20,bg="powder blue")
        FrameDetails.place(x=0,y=590,width=1530,height=210)

        xscroll=ttk.Scrollbar(FrameDetails,orient=HORIZONTAL)
        yscroll=ttk.Scrollbar(FrameDetails,orient=VERTICAL)
        
        self.DM=ttk.Treeview(FrameDetails,column=("user","CUSTOMERID","FIRST_NAME","LAST_NAME","MOB",
                                                            "BOOKING_TYPE","CHECK_IN","CHECK_OUT","CHARGE","QUANTITY","VERIFICATIONID"),xscrollcommand=xscroll.set,yscrollcommand=yscroll.set)
        
        xscroll.pack(side=BOTTOM,fill=X)
        yscroll.pack(side=RIGHT,fill=Y)
        
        xscroll.config(command=self.DM.xview)
        yscroll.config(command=self.DM.yview)
        
        self.DM.heading("user",text="USER")
        self.DM.heading("CUSTOMERID",text="CUSTOMERID")
        self.DM.heading("FIRST_NAME",text="FIRST_NAME")
        self.DM.heading("LAST_NAME",text="LAST_NAME")
        self.DM.heading("MOB",text="MOB")
        self.DM.heading("BOOKING_TYPE",text="BOOKING_TYPE")
        self.DM.heading("CHECK_IN",text="CHECK_IN")
        self.DM.heading("CHECK_OUT",text="CHECH_OUT")
        self.DM.heading("CHARGE",text="CHARGE")
        self.DM.heading("QUANTITY",text="QUANTITY")
        self.DM.heading("VERIFICATIONID",text="VERIFICATIONID")
        self.DM["show"]="headings"
        self.DM.pack(fill=BOTH,expand=1) 

        self.DM.column("user",width=100)
        self.DM.column("CUSTOMERID",width=100)
        self.DM.column("FIRST_NAME",width=150)
        self.DM.column("LAST_NAME",width=150)
        self.DM.column("MOB",width=100)
        self.DM.column("BOOKING_TYPE",width=100)
        self.DM.column("CHECK_IN",width=100)
        self.DM.column("CHECK_OUT",width=100)
        self.DM.column("CHARGE",width=100)
        self.DM.column("QUANTITY",width=100)
        self.DM.column("VERIFICATIONID",width=100)
        
        
        self.fatch_data()
        self.DM.bind("<ButtonRelease-1>",self.get_cursor)


    def update(self):
        conn=mysql.connector.connect(host="localhost",database="PY24",user="root",password="redhat")
        my_cursor = conn.cursor()
        my_cursor.execute("update DM set  USER Type=%s,CUSTOMERID=%s,FIRST_NAME=%s, LAST_NAME=%s, MOB=%s, BOOKING_TYPE=%s, CHECK_IN=%s, CHECK_OUT=%s, CHARGE=%s, QUANTITY=%s, TOTAL_AMOUNT=%s, VERIFICATIONID=%s",
        (self.USER_var.get(),
         self.CUSTOMERID_var.get(),
         self.FIRST_NAME_var.get(),
         self.LAST_NAME_var.get(),
         self.MOB_var.get(),
         self.BOOKING_TYPE_var.get(),
         self.CHECK_IN_var.get(),
         self.CHECK_OUT_var.get(),
         self.CHARGE_var.get(),
         self.QUANTITY_var.get(),
         self.TOTAL_AMOUNT_var.get(),
         self.VERIFICATIONID_var.get(),
        ))
        
        conn.commit()
        self.fatch_data()
        self.reset()
        conn.close()
        messagebox.showinfo("Success","Member Has Been Updated Successfully")
        
    def add_data(self):
        conn=mysql.connector.connect(host="localhost",database="PY24",user="root",password="redhat")
        my_cursor = conn.cursor()
        my_cursor.execute("insert into DM values(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)",
        (self.USER_var.get(),
         self.CUSTOMERID_var.get(),
         self.FIRST_NAME_var.get(),
         self.LAST_NAME_var.get(),
         self.MOB_var.get(),
         self.BOOKING_TYPE_var.get(),
         self.CHECK_IN_var.get(),
         self.CHECK_OUT_var.get(),
         self.CHARGE_var.get(),
         self.QUANTITY_var.get(),
         self.TOTAL_AMOUNT_var.get(),
         self.VERIFICATIONID_var.get(),
        ))
        
        conn.commit()
        self.fatch_data()
        self.reset()
        conn.close()
        messagebox.showinfo("Success","Member Has Been Updated Successfully")


    def fatch_data(self):
        conn=mysql.connector.connect(host="localhost",database="PY24",user="root",password="redhat")
        my_cursor = conn.cursor()
        my_cursor.execute("select * from DM")
        rows=my_cursor.fetchall()
        
        if len(rows)!=0:
            self.DM.delete(*self.DM.get_children())
            for i in rows:
                self.DM.insert("",END,values=i)
            conn.commit()
        conn.close()

    
    def get_cursor(self,event=""):
       
      cursor_row=self.DM.focus()
      content=self.DM.item(cursor_row)
      row=content["values"]
        

      self.USER_var.set(row[0]),
      self.CUSTOMERID_var.set(row[1]),
      self.FIRST_NAME_var.set(row[2]),
      self.LAST_NAME_var.set(row[3]),
      self.MOB_var.set(row[4]),
      self.BOOKING_TYPE_var.set(row[5]),
      self.CHECK_IN_var.set(row[6]),
      self.CHECK_OUT_var.set(row[7]),
      self.CHARGE_var.set(row[8]),
      self.QUANTITY_var.set(row[9]),
      self.TOTAL_AMOUNT_var.set(row[10]),
      self.VERIFICATIONID_var.set(row[11])


    def showData(self):
        self.txtBox.insert(END,"USER type\t\t"+self.USER_var.get()+"\n")
        self.txtBox.insert(END,"CUSTOMERID\t\t"+self.CUSTOMERID_var.get()+"\n")
        self.txtBox.insert(END,"FIRST_NAME\t\t"+self.FIRST_NAME_var.get()+"\n")                
        self.txtBox.insert(END,"LAST_NAME\t\t"+self.LAST_NAME_var.get()+"\n")
        self.txtBox.insert(END,"MOB\t\t"+self.MOB_var.get()+"\n")
        self.txtBox.insert(END,"BOOKING_TYPE\t\t"+self.BOOKING_TYPE_var.get()+"\n")
        self.txtBox.insert(END,"CHECK_IN\t\t"+self.CHECK_IN_var.get()+"\n")                
        self.txtBox.insert(END,"CHECK_OUT\t\t"+self.CHECK_OUT_var.get()+"\n")                   
        self.txtBox.insert(END,"CHARGE\t\t"+self.CHARGE_var.get()+"\n")                
        self.txtBox.insert(END,"QUANTITY\t\t"+self.QUANTITY_var.get()+"\n")                   
        self.txtBox.insert(END,"TOTAL_AMOUNT\t\t"+self.TOTAL_AMOUNT_var.get()+"\n")                 
        self.txtBox.insert(END,"VERIFICATIONID\t\t"+self.VERIFICATIONID_var.get()+"\n")


    def reset(self):
        
         self.USER_var.set(""),
         self.CUSTOMERID_var.set(""),
         self.FIRST_NAME_var.set(""),
         self.LAST_NAME_var.set(""),
         self.MOB_var.set(""),
         self.BOOKING_TYPE_var.set(""),
         self.CHECK_IN_var.set(""),
         self.CHECK_OUT_var.set(""),
         self.CHARGE_var.set(""),
         self.QUANTITY_var.set(""),
         self.TOTAL_AMOUNT_var.set(""),
         self.VERIFICATIONID_var.set("")
         self.txtBox.delete("1.0",END)


    def iExit(self):
        iExit=Tk.messagebox.askyesno("Hotel Management System","Do you want to exit")
        if iExit>0:
            self.root.destroy()
            return
            
    
    def delete(self):
        if self.CUSTOMERID_var.get()=="":
            messagebox.showerror("Error","First Select the USER")
        else:
            conn=mysql.connector.connect(host="localhost",database="PY24",user="root",password="redhat")
            my_cursor = conn.cursor()
            query="delete from DM where VERIFICATIONID=%s"
            value=(self.VERIFICATIONID_var.get(),)
            my_cursor.execute(query,value)
            
            conn.commit()
            self.fatch_data()
            self.reset()
            conn.close()
            messagebox.showinfo("Success","USER Has Been Deleted Successfully")



if __name__=="__main__":
    root=Tk()
    obj=HotelManagementSystem(root)
    root.mainloop()
            
```
